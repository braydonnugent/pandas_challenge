```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("../Starter_Code/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("../Starter_Code/PyCitySchools/Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
# Define LGA variables

students = school_data_complete['student_name']
schools = school_data_complete['school_name']
school_types = school_data_complete['type']
budget = school_data_complete['budget']
maths_score = school_data_complete['maths_score']
reading_score = school_data_complete['reading_score']
student_id = school_data_complete['Student ID']
year = school_data_complete['year']
```

```python
last_row_complete = school_data_complete.iloc[-1, 0]
last_row_schools = school_data.iloc[-1, 0]
total_budget = 0
total_maths = 0
total_reading = 0
maths_passing = 0
reading_passing = 0
overall_passing = 0
total_schools = schools.nunique()
total_students = students.shape[0]

for i in range(last_row_schools + 1):
    budget_value = school_data.iloc[i]['budget']
    total_budget += budget_value

for i in range(last_row_complete):
    ind_maths = school_data_complete.iloc[i]['maths_score']
    total_maths += ind_maths
    ind_reading = school_data_complete.iloc[i]['reading_score']
    total_reading += ind_reading

    if school_data_complete.iloc[i]['maths_score'] >= 50:
        maths_passing = maths_passing + 1

    if school_data_complete.iloc[i]['reading_score'] >= 50:
        reading_passing = reading_passing + 1

    if school_data_complete.iloc[i]['maths_score'] >= 50 and school_data_complete.iloc[i]['reading_score'] >= 50:
        overall_passing = overall_passing + 1


average_maths = f"{total_maths/last_row_complete:.2f}"
average_reading = f"{total_reading/last_row_complete:.2f}"
maths_passing_percent = f"{maths_passing/last_row_complete*100:.2f}%"
reading_passing_percent = f"{reading_passing/last_row_complete*100:.2f}%"
overall_passing_percent = f"{overall_passing/last_row_complete*100:.2f}%"
total_budget = f"${total_budget}"
```

```python
# Create Data Frame
lga_columns1 = ['Total Schools', 'Total Students', 'Total Budget', 'Average Maths Scores','Average Reading Scores', 'Percentage Passing Maths', 'Percentage Passing Reading', 'Percentage Overall Passing']
lga_columns2 = [total_schools, total_students, total_budget, average_maths, average_reading, maths_passing_percent, reading_passing_percent, overall_passing_percent]
area_summary = pd.DataFrame(columns=[lga_columns1, lga_columns2])

area_summary
```

```python
# Define School Summary Variables
school_list = []
prev_school = None

for index, row in school_data_complete.iterrows():
    if row['school_name'] != prev_school:
        school_list.append([row['school_name']])
        prev_school = row['school_name']


school_list = list(school_list)
school_list = [item[0] for item in school_list]
```

```python
# Variables
summary_dict = []
z = 0
students_dict = 0
budget_dict = 0
m_score_dict = 0
r_score_dict = 0
pass_m_count = 0
pass_r_count = 0
pass_o_count = 0

for i in range(last_row_complete):
    if school_list[z] == schools[i]:
        schools_dict = schools[i]
        type_dict = school_types[i]
        students_dict += 1
        budget_dict = budget[i]
        m_score_dict += maths_score[i]
        r_score_dict += reading_score[i]
        if maths_score[i] >= 50:
            pass_m_count += 1
        if reading_score[i] >= 50:
            pass_r_count += 1
        if maths_score[i] >= 50 and reading_score[i] >= 50:
            pass_o_count += 1

    elif school_list[z] != schools[i]:
        s_budget_dict = budget_dict/students_dict
        avr_mscore_dict = m_score_dict/students_dict
        avr_rscore_dict = r_score_dict/students_dict
        percent_mpass_dict = pass_m_count/students_dict*100
        percent_rpass_dict = pass_r_count/students_dict*100
        percent_opass_dict = pass_o_count/students_dict*100

        school_data = [school_list[z], type_dict, students_dict, budget_dict, s_budget_dict, avr_mscore_dict, avr_rscore_dict, percent_mpass_dict, percent_rpass_dict, percent_opass_dict]
        summary_dict.append(school_data)

        students_dict = 1
        type_dict = school_types[i]
        budget_dict = budget[i]
        m_score_dict = maths_score[i]
        r_score_dict = reading_score[i]
        pass_m_count = 1 if maths_score[i] >= 50 else 0
        pass_r_count = 1 if reading_score[i] >= 50 else 0
        pass_o_count = 1 if maths_score[i] >= 50 and reading_score[i] >= 50 else 0

        z += 1


s_budget_dict = budget_dict/students_dict
avr_mscore_dict = m_score_dict/students_dict
avr_rscore_dict = r_score_dict/students_dict
percent_mpass_dict = pass_m_count/students_dict*100
percent_rpass_dict = pass_r_count/students_dict*100
percent_opass_dict = pass_o_count/students_dict*100


school_data = [school_list[z], type_dict, students_dict, budget_dict, s_budget_dict,avr_mscore_dict, avr_rscore_dict, percent_mpass_dict, percent_rpass_dict, percent_opass_dict]
summary_dict.append(school_data)
```

```python
school_summary_df = pd.DataFrame(summary_dict, columns=["School Name", "School Type", "Total Students", "Total Budget", "Budget per Student","Average Math Score", "Average Reading Score", "Percentage Passing Math","Percentage Passing Reading", "Percentage Overall Passing"])
per_school_summary = school_summary_df.set_index('School Name')

per_school_summary
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-$630", "$630-$645", "$645-$680"]

school_summary_df['Spending Range'] = pd.cut(school_summary_df['Budget per Student'], bins=spending_bins, labels=labels)
grouped_data = school_summary_df.groupby('Spending Range')[['Average Math Score', 'Average Reading Score', 'Percentage Passing Math', 'Percentage Passing Reading', 'Percentage Overall Passing']].mean()
spending_summary = pd.DataFrame(grouped_data, columns=['Average Math Score', 'Average Reading Score', 'Percentage Passing Math', 'Percentage Passing Reading', 'Percentage Overall Passing'])

spending_summary
```
